## The VGG16 model was developed by the Visual Graphics Group

```python
from keras.applications.vgg16 import VGG16

model = VGG16()
```

```python
model.summary()
```

```python
from keras.preprocessing.image import load_img
from keras.applications.imagenet_utils import preprocess_input
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread

image = load_img('redmi-a5-118.jpeg', target_size=(224,224))

image = np.array(image)

image =image.reshape((1,image.shape[0],image.shape[1],image.shape[2]))

image = preprocess_input(image)
image

my_image = imread('redmi-a5-118.jpeg')
plt.imshow(my_image)
plt.axis('off')  # optional: hides axes
plt.show()
```

```python
yhat = model.predict(image)
```

```python
from keras.applications.vgg16 import preprocess_input, decode_predictions
labels = decode_predictions(yhat)
labels
```
